```python
gestures = ['basta', 'messi', 'ok', 'cosat', 'combi', 'chevu', 'perfe', 'nonce']
joint_type = 'HandRight'

n_figs = len(gestures)
figs, axs = plt.subplots(nrows=n_figs, ncols=3, figsize=(20,30))
i=0
df_sequence = df.groupby('sample_id')             

for gesture in gestures:
    df_ = df.groupby('gesture').get_group(gesture).groupby('JointType').get_group(joint_type).groupby('sample_id')


    for name, group in df_:
        axs[i][0].plot(group.frame - group.frame.min(), group.y_pix - df_sequence.get_group(name).y_pix.mean() , 'b-', alpha=0.2)
        axs[i][1].plot(group.frame - group.frame.min(), group.x_pix - df_sequence.get_group(name).x_pix.mean(), 'b-', alpha=0.2)
        
        axs[i][2].plot(group.frame - group.frame.min(),
                 (group.x_pix - df_sequence.get_group(name).x_pix.mean()) * \
                 (group.y_pix - df_sequence.get_group(name).y_pix.mean()), 'b-', alpha=0.2)
    i += 1
```

```python
def aggregate_skelettion(file_names=['training1'], agg_functions=['mean']):
    X = DataFrame()
    
    for file_name in file_names:
        df = preprocessed_skeleton(file_name)
        df.drop('frame')
        df = df.groupby(['sample_id', 'gesture', 'JointType']).agg(agg_functions).unstack('JointType')
        X = pd.concat([X, df])
    y = np.array([gesture for (sample_id, gesture) in X.index])
    return X, y    
```
